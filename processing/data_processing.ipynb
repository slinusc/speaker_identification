```python
def segment_npy_files(source_directory, target_directory, segment_length, overlap=0):
    os.makedirs(target_directory, exist_ok=True)
    print(f"Processing files in {source_directory}")
    for filename in os.listdir(source_directory):
        if filename.endswith(".npy"):
            file_path = os.path.join(source_directory, filename)
            data = np.load(file_path)
            print(f"Loaded {filename} with data shape: {data.shape}")

            # Assume the segments are to be created along the second dimension
            step_size = segment_length - overlap
            num_segments = (data.shape[1] - overlap) // step_size
            print(f"Creating {num_segments} segments from {filename}")

            segments = []
            for i in range(num_segments):
                if i * step_size + segment_length <= data.shape[1]:
                    # Segment along the second dimension
                    segment = data[:, i * step_size: i * step_size + segment_length, :]
                    segments.append(segment)

            for idx, segment in enumerate(segments):
                segment_filename = f"{os.path.splitext(filename)[0]}_seg{idx}.npy"
                segment_path = os.path.join(target_directory, segment_filename)
                np.save(segment_path, segment)
                print(f"Segment saved: {segment_path}")

# Example usage
source_dir =  '/content/drive/My Drive/NLP/clips__test/Speaker1/layer_1'
target_dir = '/content/drive/My Drive/NLP/clips__test/Speaker1/layer_1_processed'
segment_len = 100

segment_npy_files(source_dir, target_dir, segment_len)
```

```python
# PROCESS ALL FILES FOR ALL SPEAKERS
import os
import numpy as np

def segment_npy_files(base_directory, layer_name, segment_length, overlap=0):
    for speaker_idx in range(1, 21):  # Assuming speaker folders are named Speaker1 to Speaker20
        source_directory = os.path.join(base_directory, f"Speaker{speaker_idx}", layer_name)
        target_directory = os.path.join(base_directory, f"Speaker{speaker_idx}", f"{layer_name}_processed")
        os.makedirs(target_directory, exist_ok=True)

        print(f"Processing files in {source_directory}")
        for filename in os.listdir(source_directory):
            if filename.endswith(".npy"):
                file_path = os.path.join(source_directory, filename)
                data = np.load(file_path)
                print(f"Loaded {filename} with data shape: {data.shape}")

                # Segment along the second dimension
                step_size = segment_length - overlap
                num_segments = (data.shape[1] - overlap) // step_size
                print(f"Creating {num_segments} segments from {filename}")

                segments = []
                for i in range(num_segments):
                    if i * step_size + segment_length <= data.shape[1]:
                        segment = data[:, i * step_size: i * step_size + segment_length, :]
                        segments.append(segment)

                for idx, segment in enumerate(segments):
                    segment_filename = f"{os.path.splitext(filename)[0]}_seg{idx}.npy"
                    segment_path = os.path.join(target_directory, segment_filename)
                    np.save(segment_path, segment)
                    print(f"Segment saved: {segment_path}")

# Example usage parameters
base_directory = '/content/drive/My Drive/NLP/clips__test'
layer_name = 'layer_24'
segment_length = 100
overlap = 10

# Call the function
segment_npy_files(base_directory, layer_name, segment_length, overlap)

```
